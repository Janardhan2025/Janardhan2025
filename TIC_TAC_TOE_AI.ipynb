```python
import math

board=[' ' for _ in range(9)]

def print_board(board):
    print()
    for i in range(3):
        print(' ' + ' | '.join(board[i*3:(i+1)*3]))
        if i<2:print("---+---+---")
    print()

def available_moves(board):
    return [i for i,s in enumerate(board) if s==' ']

def winner(board):
    wins=[(0,1,2),(3,4,5),(6,7,8),(0,3,6),(1,4,7),(2,5,8),(0,4,8),(2,4,6)]
    for a,b,c in wins:
        if board[a]==board[b]==board[c] and board[a]!=' ':return board[a]
    return None

def is_full(board):
    return ' ' not in board

def minimax(board,depth,is_max,alpha,beta):
    win=winner(board)
    if win=='O':return 1
    if win=='X':return -1
    if is_full(board):return 0
    if is_max:
        best=-math.inf
        for m in available_moves(board):
            board[m]='O'
            val=minimax(board,depth+1,False,alpha,beta)
            board[m]=' '
            best=max(best,val)
            alpha=max(alpha,val)
            if beta<=alpha:break
        return best
    else:
        best=math.inf
        for m in available_moves(board):
            board[m]='X'
            val=minimax(board,depth+1,True,alpha,beta)
            board[m]=' '
            best=min(best,val)
            beta=min(beta,val)
            if beta<=alpha:break
        return best

def ai_move(board):
    best=-math.inf
    move=None
    for i in available_moves(board):
        board[i]='O'
        val=minimax(board,0,False,-math.inf,math.inf)
        board[i]=' '
        if val>best:
            best=val;move=i
    return move

def play_game():
    print("Welcome to Tic-Tac-Toe!")
    print("You are X, AI is O.")
    print_board(board)
    while True:
        while True:
            try:
                m=int(input("Enter move (1-9): "))-1
                if m in available_moves(board):
                    board[m]='X';break
                else:print("Invalid.")
            except:print("Enter 1-9.")
        print_board(board)
        if winner(board):print("You win!");break
        if is_full(board):print("Draw!");break
        print("AI is thinking...")
        m=ai_move(board)
        board[m]='O'
        print_board(board)
        if winner(board):print("AI wins!");break
        if is_full(board):print("Draw!");break

play_game()
```
