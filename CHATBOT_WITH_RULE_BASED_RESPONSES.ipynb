```python
# Simple Rule-Based Safety Chatbot (No Weather)

print("ü§ñ SafetyBot: Hello! I am your safety assistant.")
print("Type 'bye' to exit.\n")

while True:
    user_input = input("You: ").lower()

    # Exit
    if "bye" in user_input:
        print("ü§ñ SafetyBot: Stay safe! Goodbye üôÇ")
        break

    # Health safety
    elif "health" in user_input or "health safety" in user_input:
        print("ü©∫ Health Safety Tips:")
        print("- Wash your hands regularly.")
        print("- Drink enough water daily.")
        print("- Get 7‚Äì8 hours of sleep.")
        print("- Avoid junk food; eat more fruits & vegetables.")
        print("- Do regular exercise or yoga.")

    # Car safety
    elif "car" in user_input or "drive" in user_input or "driving" in user_input:
        print("üöó Car/Driving Safety Tips:")
        print("- Always wear your seatbelt.")
        print("- Do not use your phone while driving.")
        print("- Follow speed limits.")
        print("- Maintain proper tire pressure.")
        print("- Keep a first-aid kit in the car.")

    # Home safety
    elif "home" in user_input:
        print("üè† Home Safety Tips:")
        print("- Turn off gas and electrical appliances when not in use.")
        print("- Install smoke detectors.")
        print("- Keep emergency numbers ready.")
        print("- Lock doors and windows properly.")
        print("- Keep floors dry to avoid slipping.")

    # Outdoor safety
    elif "outdoor" in user_input or "outside" in user_input:
        print("üå≥ Outdoor Safety Tips:")
        print("- Avoid isolated places at night.")
        print("- Stay hydrated when going out.")
        print("- Use sunscreen during daytime.")
        print("- Keep your phone charged.")
        print("- Inform someone if travelling alone.")

    # Emergency safety
    elif "emergency" in user_input:
        print("üö® Emergency Safety Tips:")
        print("- Stay calm and think clearly.")
        print("- Call emergency numbers immediately.")
        print("- Follow evacuation instructions if given.")
        print("- Help others if safe to do so.")
        print("- Keep an emergency kit ready.")

    # Greetings
    elif "hello" in user_input or "hi" in user_input:
        print("ü§ñ SafetyBot: Hello! Ask me about health safety, car safety, home safety, outdoor safety, or emergency tips.")

    # Default response
    else:
        print("ü§ñ SafetyBot: I can give safety advice. Try: 'health safety', 'car safety', 'home safety', or 'emergency safety'.")
```
